```python
#1 sentiment-analysis

from transformers import pipeline

# Sentiment analysis pipeline
sentiment = pipeline("sentiment-analysis")

print(sentiment("I love HuggingFace!"))
print(sentiment("This is the worst movie I have ever seen."))
```

```python
#2 question-answering

qa = pipeline("question-answering")

context = """Hugging Face is an open-source AI company that created the transformers library.
It is widely used for NLP tasks like sentiment analysis, summarization, and translation."""


answer = qa({ "question" : "Who created the transformers library?" ,
     "context" : context
 })
print(answer)
print(answer.get("answer"))



```

```python
# 3 Summarization
summarizer = pipeline("summarization")

article = """Hugging Face is a company that provides machine learning tools.
Its most popular library is transformers which allows developers to use
pretrained models for natural language processing tasks like sentiment analysis,
translation, summarization, and text generation."""

final_summrization = summarizer(article, max_length=40, min_length=10, do_sample=False)

print(final_summrization)
```

```python
#4 zero-shot-classification

classifier = pipeline("zero-shot-classification")

text = "I love learning AI with HuggingFace."
labels = ["education", "politics", "sports"]

print(classifier(text, candidate_labels=labels))
```

```python
# 5 Name-Entit-Recognization

ner = pipeline("ner", grouped_entities=True)

print(ner("Elon Musk founded SpaceX in California."))
```

```python
# 6 translation

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-hi")

print(translator("HuggingFace is awesome!"))
```

```python
#7 text-generation

pipeline = pipeline(task="text-generation", model="google/gemma-2-2b", device=0)
pipeline("what is artificial intelligence?", max_new_tokens=50, do_sample=True, temperature=0.9)
```

```python
#8 fill mask

unmasker = pipeline("fill-mask")

print(unmasker("HuggingFace is a <mask> library.")[0])
```
